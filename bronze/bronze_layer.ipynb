```python
dbutils.widgets.text("file_name", "")
```

```python
p_file_name = dbutils.widgets.get("file_name")
```

```python
df = spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format", "csv")\
    .option("cloudFiles.schemaLocation", f"abfss://bronze@ecommercestoragev2.dfs.core.windows.net/checkpoint_{p_file_name}")\
    .load(f"abfss://source@ecommercestoragev2.dfs.core.windows.net/{p_file_name}")
```

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation", f"abfss://bronze@ecommercestoragev2.dfs.core.windows.net/checkpoint_{p_file_name}")\
    .option("path", f"abfss://bronze@ecommercestoragev2.dfs.core.windows.net/{p_file_name}/")\
    .trigger(once=True)\
    .start()
```
